## 4. 워크플로우 설명

### 멀티턴 프로세스 흐름

1. **Suggestion Process (제안 프로세스)**
   - JD 검색: 관련 직무 설명서 정보 검색
   - 최신 정보 검색: 최신 트렌드 정보 검색
   - 문서 수정: 검색 결과를 바탕으로 문서 수정

2. **Review Process (검토 프로세스)**
   - 문법 검사: 수정된 문서의 문법 오류 검사
   - ATS 분석: 지원자 추적 시스템 최적화 분석
   - JD 리포트: 직무 요구사항 충족 여부 리포트 제공

3. **Decision (결정)**
   - 검토 완료 여부 결정: 문서가 충분히 개선되었는지 판단
   - 추가 개선 필요 시: 피드백과 함께 Suggestion Process로 돌아감
   - 검토 완료 시: 프로세스 종료

이 워크플로우는 문서가 충분히 개선될 때까지 반복적으로 실행됩니다.
